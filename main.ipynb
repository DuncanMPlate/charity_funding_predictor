## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
# While attempting to acieve above 75% accuracy, I went back and removed 'USE_CASE' and 'ORGANIZATION' columns. These changes resulted in a reduction of accuracy by 1%
# After dropping those columns, I tried to keep the NAME column resulted in an evaluation of Loss: 0.7053233981132507, Accuracy: 0.7162682414054871
dropped_df = application_df.drop(columns=['EIN', 'NAME'], axis=1)
dropped_df
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
dropped_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
dropped_df['APPLICATION_TYPE'].value_counts()
```

```python
#from sklearn.metrics import roc_curve, auc
```

```python
#In the endeavor of trying to get the damn model to work, I went back and added T9 and T13 to the non-OTHER field. 
application_types_to_replace = ['T12', 'T2', 'T14', 'T25', 'T15', 'T25', 'T29', 'T17']
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE



# Replace in dataframe
for app in application_types_to_replace:
    dropped_df['APPLICATION_TYPE'] = dropped_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
dropped_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
dropped_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
dropped_df['CLASSIFICATION'].value_counts().loc[lambda x: x>1]
```

```python
classifications_to_replace = ['C7000', 'C1700', 'C4000', 'C5000', 'C1270', 'C2700', 'C2800', 'C7100', 'C1300', 'C1280', 'C1230', 'C1400', 'C2300', 'C7200', 'C1240', 'C8000', 'C7120', 'C1500', 'C1800', 'C6000', 'C1250', 'C8200', 'C1238', 'C1278', 'C1237', 'C1235', 'C7210', 'C2400', 'C4100', 'C1720', 'C1257', 'C1600', 'C0','C2710', 'C1260', 'C1234', 'C1267', 'C1246', 'C1256', 'C3200','C2150','C2190','C1283','C4500','C4120','C1236','C2500','C2600','C5200','C1248','C3700','C2570','C4200','C1820','C2170','C2561','C1728','C2380','C1245','C8210','C6100','C1732','C1900','C1570','C1370','C1580']
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE

# Replace in dataframe
for cls in classifications_to_replace:
    dropped_df['CLASSIFICATION'] = dropped_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
dropped_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
dummies_df = pd.get_dummies(dropped_df)
dummies_df
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y= dummies_df['IS_SUCCESSFUL']
X= dummies_df.drop(columns='IS_SUCCESSFUL', axis=1)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
#  YOUR CODE GOES HERE
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#inital nn_model, could not provide any growth to its accuracy over testing time, staying at about 73.5%

nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(units=16, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=16, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=16, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Compile the Sequential model together and customize metrics
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)

# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
